```python
df_com = df_com[df_com['prod'].isin({'Credit Card', 'Cash Loan', 'Debit Card', })]
```

```python
df_com['channel_prod_comm'] = df_com['channel'] + '_' + df_com['prod']
df_com.head()
```

```python
df_grouped = df_com.groupby(['client_id', 'prod']).agg('sum').reset_index()
df_grouped.head()
```

```python
df_features_pull = []

for values in ['agr_flg', 'otkaz', 'dumaet', 'ring_up_flg', 'not_ring_up_flg', 'count_comm']:
    df_features_ = pd.pivot_table(
        df_grouped,
        values=values,
        index='client_id', 
        columns='prod',
    )
    df_features_.columns = df_features_.columns.str.replace(' ', '_')
    df_features_ = df_features_.add_prefix('prod_' + values + '_')
    df_features_pull.append(df_features_)

df_features = pd.concat(df_features_pull, axis=1)
df_features.head()
```

```python
counts = df_features.isna().sum(axis=0) / df_features.shape[0]
counts = counts[counts >= 0.95]
counts.index
```

```python
df_features = df_features.fillna(0).astype(int)
```
